## 반지름이 10,15,20인 원의 면적을 구하시오 

```r
# 함수 생성 -> 인자값으로 반지름과 pi값을 받는다 
circle1  <- function(x,pi){
    (x^2)*3.14 # 원의 넓이를 구하는 식을 다음과 같이 정의(x = 반지름, pi는 3.14)
}
```

```r
circle1(10,3.14)
circle1(15,3.14)
circle1(20,3.14)
```

```r
# 101~200의 값으로 구성된 벡터 d를 생성하시오
d <- 101:200
# d에 어떤 값이 저장되어있는지 확인하시오 
d
# d에서 10번째 값은 무엇인가?
d[10]
# d에서 뒤에서 10개의 값을 잘라내어 보이시오 
d[90:100]
# d에서 짝수만 출력하시오 
d[seq(2,100,2)]
# d에서 앞에서 20개의 값을 잘라내어 d.20에 저장하시오. d.20의 값을 보이시오
d.20 <- d[1:20]
d.20
# d.20에서 5번째 값을 제외한 나머지 값들을 보이시오
d.20[-5]
# d.20에서 5,7,9 번째 값을 제외한 나머지 값을 보이시오
d.20[c(-5,-7,-9)]
```

## 데이터 벡터에 적용 가능한 함수들 
### sum, mean, median, max, min, sort, range 등등

```r
# [연습4]
```

```r
# d1,d2가 다음과 같을 때 질문에 답하시오 
d1 <- 1:50
d2 <- 51:100
# d1,d2의 값을 보이시오
d1
d2
# d1 +d2, d2-d1, d1*d2,d2/d1의 결과를 각각 보이시오
d1 + d2
d2 - d1 
d1 * d2
d2 / d1
# d1,d2의 값들의 합을 각각 보이시오 
sum(d1)
sum(d2)
# d1,d2에 있는 모든 값들의 합을 보이시오
sum(d1+d2)
# d2에서 가장 큰 값과 가장 작은 값을 보이시오 
max(d2)
min(d2)
# d2와 d1의 값들의 평균값을 각각 구하고 두 평균의 차이를 보이시오 
mean(d2) - mean(d1)
#d1의 값들을 큰수에서 작은 수 순으로 정렬하여 보이시오
sort(d1,decreasing = TRUE)
# d1과 d2에서 큰수 순으로 각각 10개씩 추출하여 d3에 저장하시오(결과적으로 d3에너느 20개의 수가 저장)
d1 <- sort(d1,decreasing = TRUE)
d2 <- sort(d2,decreasing = TRUE)
d3 <- c(d1[1:10],d2[1:10])
d3
```

## 논리 연산자: <,<=.>,>=,==,!=,|,&

# [연습5]

```r
# 다음과 같이 벡터 v1을 생성하시오 
v1 <- 51:90
# v1에서 60보다 작은 수들을 보이시오 
v1[v1<60]
# v1에서 70보다 작은 수들은 몇개인지 보이시오
length(v1[v1<70])
# v1에서 65보다 큰수들의 합을 보이시오 
sum(v1[v1>65])
# v1에서 60보다 크고 73보다 작은 수들을 보이시오 
v1[v1>60 & v1<73]
# v1에서 65보다 작거나 80보다 큰 수들을 보이시오
v1[v1<65 | v1>80]
# v1에서 7로 나눈 나머지가 3인 숫자들을 보이시오 
v1[v1%%7 == 3]
# v1에서 짝수들의 합계를 보이시오 
sum(v1[v1%%2 == 0])
# v1에서 홀수이거나 80보다 큰수를 보이시오
v1[v1%%2 != 0 | v1>80]
#v1에서 3과 5의 공배수를 보이시오 
v1[v1%%3 == 0 & v1%%5 == 0]
```
